```python
from google.colab import drive
from ultralytics import YOLO
import glob
```

```python
drive.mount('/content/drive')
path = '/content/drive/My Drive/AirplanesDetection/dataset.yaml'
```

```python
model = YOLO("yolov8m.pt")

model.train(data = path,
            epochs = 50,
            imgsz = 960,
            device = 0,
            verbose = True,
            workers = 0)
print("Training completed.")
```

```python
model.export(format = "onnx")
```

```python
model = YOLO('/content/drive/My Drive/AirplanesDetection/best.onnx', task = 'detect')
```

```python
results = model('/content/drive/My Drive/AirplanesDetection/images_test/')
for result in results:
  result.show()
  print("")
```
